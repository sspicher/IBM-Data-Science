### My code will print out the mean of a list of n random numbers between 0 and 1

````python
import random
def mean(n):
    mylist=[]
    for i in range(0,n):
        x = random.random()
        mylist.append(x)
    mean = sum(mylist)/n
    return mean

mean(50)
````
